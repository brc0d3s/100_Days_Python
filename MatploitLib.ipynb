# What is Matplotlib?

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.array([0,5,20,45])
y = np.array([0,3,5,11])
plt.plot(x,y)
plt.show()
```

### Plotting Without Line

```python
plt.plot(x,y, "o")
```

```python
plt.plot(y)
```

## Matplotlib Markers

```python
plt.plot(x,y, marker = "o")
```

```python
plt.plot(x,y, "x:r")
```
