# What is Matplotlib?

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.array([0,5,20,45])
y = np.array([0,3,5,11])
plt.plot(x,y)
plt.show()
```

### Plotting Without Line

```python
plt.plot(x,y, "o")
```

```python
plt.plot(y)
```

## Matplotlib Markers

```python
plt.plot(x,y, marker = "o")
```

```python
plt.plot(x,y, "x:r")
```

```python
start = np.array([0,1,3])
stop = np.array([0,8,10])

plt.plot(start,stop,marker = "o")
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x,y, linestyle = "dotted",c = "b")
plt.title("Sports data")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.show()
```

## Matplotlib Adding Grid Lines

```python
speed = np.array([45,67,53,87,23,13,534])
time = np.array([23,56,99,87,67,12,435])

plt.plot(speed,time, color = "blue")
plt.title("Speed Against Time")
plt.xlabel("Speed")
plt.ylabel("Time")
plt.grid()
plt.show()
```
