```python
def getGrade(m):
    if(70 <= m < 100):
        grade = "A"
    elif(60 <= m < 70):
        grade = "B"
    elif(50 <= m < 60):
        grade = "C"
    elif(40 <= m < 50):
        grade = "D"
    elif(0 <= m < 40):
        grade = "E"
    else:
        grade = "Invalid"
    return grade

marks = int(input("Enter Your Marks: "))
print(f"Your grade is: {getGrade(marks)}")
```

```python
# Initial list with None for missing marks
scores = [61, 59, 72, 56, None, 38, None]

units = ["SOC 100", "SOC 102", "SOC 104", "SOC 105", "SOC 107", "SOC 111", "SOC 112"]

print("Intial Transcript")
for i in range(len(units)):
    print(f"{units[i]} : {scores[i]}")
print(f"\nTotal number of units: {len(units)}")
```

```python
# ii. Update missing scores for SOC 107 and SOC 112
scores[4] = 64 
scores[6] = 53  

print("\nUpdated Transcript (after filling missing scores):")
for i in range(len(units)):
    print(f"{units[i]}: {scores[i]}")
```

```python
# iii. Update the score for SOC 111 after remarking
scores[5] = 46  # Update SOC 111 score to 46

print("\nUpdated Transcript (after remarking SOC 111):")
for i in range(len(units)):
    print(f"{units[i]}: {scores[i]}")
```

```python
# iv. Compute the total marks and average marks
total_marks = 0
n = len(units)

for score in scores:
    total_marks += score

print(f"Total Marks : {total_marks}")
print(f"Average Marks: {total_marks/n}")
```
